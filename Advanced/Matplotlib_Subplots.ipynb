```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fig, axs = plt.subplots(1, 2, sharex=True, sharey=True)

# Generate 100 points between 0 and 2π
x_vals = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x_vals)
y_cos = np.cos(x_vals)

axs[0].plot(x_vals, y_sin)
axs[0].set_title("Sine Curve")

axs[1].plot(x_vals, y_cos)
axs[1].set_title("Cosine Curve")

plt.tight_layout()
plt.show()
```
