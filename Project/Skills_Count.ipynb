```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_IN  = df[df['job_country'] == 'India'].copy()
df_IN
```

```python
df_IN_exploded = df_IN.explode('job_skills')
df_IN_exploded
```

```python
df_skills_count = df_IN_exploded.pivot_table(index='job_title_short',columns='job_skills', aggfunc='size', fill_value=0)
df_skills_count
```

```python
df_skills_count.loc['Total'] = df_skills_count.sum()
df_skills_count.sort_values(axis=1,ascending=False, by='Total', inplace=True)
```

```python
top_3 = df_skills_count.iloc[:, :3].head(3)
top_3
```

```python
def percentage(skill_count):
    """
    Function to calculate the percentage of each skill in the job title
    """
    return (skill_count / skill_count.sum()) * 100

# Plotting
fig, ax = plt.subplots(3, 1, figsize=(6, 8))
for index, job_title in enumerate(top_3.index):
    # Get the skill counts for this job title and sort them
    skill_counts = top_3.loc[job_title].sort_values(ascending=True)
    ax[index].barh(skill_counts.index, percentage(skill_counts.values), label=job_title)
    ax[index].set_title(job_title)
    ax[index].set_xlabel('Percentage of Skills')
    ax[index].set_xlim(0, 70)
    ax[index].set_ylabel('')

fig.suptitle('Top 3 Jobs and their Skills in India', fontsize=16)

plt.tight_layout()
```
